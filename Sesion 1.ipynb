## Definiciones básicas

Un conjunto de $n$ números reales $x$ puede representarse como un punto en el espacio de $n$ dimensiones, $\mathbb{R}^n$. Definimos el **vector** $x$ como el **segmento orientado** que une el origen de coordenadas con el punto $x$. La dirección es importante; no es lo mismo el vector $x$ que el vector $-x$.

La **suma (o diferencia)** de dos vectores $U, V$ en $\mathbb{R}^2$ se define como un nuevo cetro con componentes iguales a la suma (diferencia) de los componentes de los sumandos. Intuitivamente la suma implica traslada un vector al extremo del otro y construir una línea que va desde el origen del primero al extremo del segundo. 

Consideremos tres vectores en el plano, $\mathbb{R}^2$: $u= (3,2), v = (-1,0)$ y representemos su suma.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos los vectores
u = np.array([3, 2])
v = np.array([-1, 0])
v_sum = u + v 

# Creamos la figura
fig, ax = plt.subplots(figsize=(5,5))

# Proyeccoón del vector desde el origen
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u (3,2)')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label='v (-1,0)')
ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='b', label="v1 + v2 (1,5)")

ax.set_xlim(-3, 4)
ax.set_ylim(-3, 4)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, linestyle='--', alpha=0.6)

ax.legend()
plt.title("Vectores en R2")
plt.show()
```

El **producto de una constante por un vector**  es un nuevo vector cuyos componentes son los del vector inicial multiplicados por la constante. En estadística equivale a un cambio en las unidades de medición. 

Imaginemos una variable $x$ de ingresos que usa como unidad el dolar, pero deseamos convertirlo a pesos

Un **vector transpuesto** $X'$, de otro $X$, es un vector con los mismos componentes de $X$, pero escritos ahora en fila:

El vector columna $\mathbf{x}$ se define como:

$$
\mathbf{x} = \begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

Su transpuesta, el vector fila $\mathbf{x'}$, es:

$$
\mathbf{x}^T = \begin{bmatrix}
x_1 & x_2 & \cdots & x_n
\end{bmatrix}
$$

```python
# vector transpuesto

v = np.array([1, 2, 3])  # Vector fila

v_transpuesto = v.T
print(f'Original: \n {v} \n Transpuesta: \n {v_transpuesto}')
```

El **producto escalar o interno** de dos vetores $X,Y \in \mathbb{R}^n$, que escribimos como $X'Y$ o $Y'X$, es el escalar obtenido al sumar los productos de sus componentes.

$$X'Y = Y'X = \sum_{i=1}^{n} x_{i}y_{i}$$

```python
# Definir los vectores
v1 = np.array([2, 3])
v2 = np.array([-1, 2])

# Producto escalar
dot_product = np.dot(v1, v2)
print("Producto escalar:", dot_product) 
```

Se llamará **norma, modulo, longitud** de un vector $X$ a la raíz cuadrada del producto escalar $X'X$. Se escribe
$\|X\|$

$$\|X\| = \sqrt{X'X} = \sqrt{x_{i}y_{i} + \cdots + x_{n}^{2}}$$

* El producto escalar de dos vectores puede calcularse también como el producto de las
normas de los vectores por el coseno del ángulo que forman.

```python
v = np.array([3, 4])
bv = np.array([3,0])
hv = np.array([0,4])

fig, ax = plt.subplots(figsize=(4,4))
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='purple')
ax.plot([v[0], v[0]], [0, v[1]], color='b', linestyle='--')
ax.plot([0, v[0]], [0, 0], color='b', linestyle='--')

ax.axhline(0, color='black',linewidth=1)
ax.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.show()
```

```python
v = np.array([3, 4])
n_v =  np.linalg.norm(v)

print(f'La norma del vector {v} es {n_v}')
```

Dos vectores son ortogonales, o perpendiculares, si y sólo si su producto escalar es cero.

```python
fig, ax = plt.subplots(figsize=(4,4))

v = np.array([1,2])
u = np.array([-2,1])

ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red')
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='purple')
ax.axhline(0, color='black',linewidth=1)
ax.axvline(0, color='black',linewidth=1)

plt.xlim(-3,3)
plt.ylim(-1,3)
plt.grid(True)
plt.show()
```
